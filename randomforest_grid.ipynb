```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
#calling the parameterized function train_test_split 
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'criterion':['poisson','squared_error','friedman_mse','absolute_error'],'max_features':['auto','sqrt','log2'],'n_estimators':[10,100]}
```

```python
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
```

```python
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
```

```python
print("The Rscore value for best parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
